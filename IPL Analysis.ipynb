```python
def runs_score(batsman_name):
    batter = delivery[delivery['batsman'] == batsman_name]
    return batter.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head(1)
```

```python
runs_score('RG Sharma')
```

# Find the most destructive death over batsman in the history of IPL
# By strike rate = (No of runs/No of balls)*100
# Min 200 balls in over 16 - 20

```python
death_over = delivery[delivery['over']>15]
last_over_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = last_over_batsman > 200
batsman_list = last_over_batsman[x].index.tolist()
```

```python
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr
```
